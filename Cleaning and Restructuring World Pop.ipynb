```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
worldpop = "world_population.csv"
worldpopdf=pd.read_csv(worldpop)
worldpopdf.head()
```

```python
# first lets check for null values
worldpopdf.isnull().sum()

#good, there are no null values
```

```python
#population change between 2022 and 2020
popchange22_20 = (worldpopdf['2022 Population']- worldpopdf['2020 Population'])/worldpopdf['2020 Population']
popchange22_20rounded = popchange22_20.round(decimals=4)
popchange22_20rounded
```

```python
#pop change 22_20 to list
popchange22_20col=popchange22_20rounded.tolist()
popchange22_20col
```

```python
# create population growth col for 2020 to 2022 pop change
worldpopdf['20-22 Pop Change']=popchange22_20col
worldpopdf
```
