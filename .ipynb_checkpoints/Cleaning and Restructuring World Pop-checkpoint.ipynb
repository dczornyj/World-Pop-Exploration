```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
worldpop = "world_population.csv"
worldpopdf=pd.read_csv(worldpop)
worldpopdf.head()
```

```python
# first lets check for null values
worldpopdf.isnull().sum()

#good, there are no null values
```

```python
#population change between 2022 and 2020
popchange22_20 = (worldpopdf['2022 Population']- worldpopdf['2020 Population'])/worldpopdf['2020 Population']
popchange22_20rounded = popchange22_20.round(decimals=4)
popchange22_20rounded
```

```python
#pop change 22_20 to list
popchange22_20col=popchange22_20rounded.tolist()
popchange22_20col
```

```python
# create population growth col for 2020 to 2022 pop change
worldpopdf['20-22 Pop Change']=popchange22_20col

worldpopdf
```

```python
columns_to_drop='Growth Rate'

worldpopdf.drop(columns_to_drop, inplace=True, axis=1)
worldpopdf
```

```python
columns_renamed = {'Rank': 'World Population Rank',
                   'CCA3': 'Abbreviation',
                   'Country/Territory': 'Country/Territory',
                   'Capital': 'Capital City',
                   'Continent': 'Continent',
                   '2022 Population': '2022 Population',
                   '20-22 Pop Change': '20-22 Population Change (%)',
                   '2020 Population': '2020 Population',
                   '2015 Population': '2015 Population',
                   '2010 Population': '2010 Population',
                   '2000 Population': '2000 Population',
                   '1990 Population': '1990 Population',
                   '1980 Population': '1980 Population',
                   '1970 Population': '1970 Population',
                   'Area (km²)': 'Country Land Area (km²)',
                   'Density (per km²)': 'People per km²',
                   'World Population Percentage':'% of World Population'
}

columns_renamed

worldpopdf.rename(columns=columns_renamed,inplace=True)
worldpopdf

# worldpopdf.rename(columns_renamed,inplace=True)
# worldpopdf
```

```python
#population change between 2020 and 2015
popchange20_15 = (worldpopdf['2020 Population']- worldpopdf['2015 Population'])/worldpopdf['2015 Population']
popchange20_15rounded = popchange20_15.round(decimals=4)
popchange20_15rounded

#population change between 2015 and 2010
popchange15_10 = (worldpopdf['2015 Population']- worldpopdf['2010 Population'])/worldpopdf['2010 Population']
popchange15_10rounded = popchange15_10.round(decimals=4)
popchange15_10rounded

#population change between 2010 and 2000
popchange10_00 = (worldpopdf['2010 Population']- worldpopdf['2000 Population'])/worldpopdf['2000 Population']
popchange10_00rounded = popchange10_00.round(decimals=4)
popchange10_00rounded

#population change between 2000 and 1990
popchange00_90 = (worldpopdf['2000 Population']- worldpopdf['1990 Population'])/worldpopdf['1990 Population']
popchange00_90rounded = popchange00_90.round(decimals=4)
popchange00_90rounded

#population change between 1990 and 1980
popchange90_80 = (worldpopdf['1990 Population']- worldpopdf['1980 Population'])/worldpopdf['1980 Population']
popchange90_80rounded = popchange90_80.round(decimals=4)
popchange90_80rounded

#population change between 1980 and 1970
popchange80_70 = (worldpopdf['1980 Population']- worldpopdf['1970 Population'])/worldpopdf['1970 Population']
popchange80_70rounded = popchange80_70.round(decimals=4)
popchange80_70rounded
```

```python
#Add each new decade increment population change percentage to a list format
new_20_15_change_col= popchange20_15rounded.tolist()

new_15_10_change_col= popchange15_10rounded.tolist()

new_10_00_change_col=popchange10_00rounded.tolist()

new_00_90_change_col=popchange00_90rounded.tolist()

new_90_80_change_col=popchange90_80rounded.tolist()

new_80_70_change_col=popchange80_70rounded.tolist()
```

```python
#append newly create lists to our dataframe so that they each make a new column
worldpopdf['2015-2020 Pop Change (%)']=new_20_15_change_col
worldpopdf['2010-2015 Pop Change (%)']=new_15_10_change_col
worldpopdf['2000-2010 Pop Change (%)']=new_10_00_change_col
worldpopdf['1990-2000 Pop Change (%)']=new_00_90_change_col
worldpopdf['1980-1990 Pop Change (%)']=new_90_80_change_col
worldpopdf['1970-1980 Pop Change (%)']=new_80_70_change_col
```
